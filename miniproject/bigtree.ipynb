```python
class BgNode:
    def __init__(self,d,l,r,n):
        self.data = d
        self.left = l
        self.right = r
        self.mid = n
        self.mult = 0
          
    # prints the node and all its children in a string
    def __str__(self):  
        st = "("+str(self.data)+", "+str(self.mult)+") -> ["
        if self.left != None:
            st += str(self.left)
        else: st += "None"
        if self.mid != None:
            st += ", "+str(self.mid)
        else: st += ", None"
        if self.right != None:
            st += ", "+str(self.right)
        else: st += ", None"
        return st + "]"


class BigTree:        
    def __init__(self):
        self.root = None
        self.size = 0

    def __str__(self):
        return str(self.root)

    def add(self,st):
        if st == "":
            return None
        if self.root == None:
            self.root = BgNode(st[0],None,None,None)
        ptr = self.root
        for i in range(len(st)):
            d = st[i]
            while True:
                if d == ptr.data:
                    break
                elif d < ptr.data:
                    if ptr.left == None:
                        ptr.left = BgNode(d,None,None,None)
                    ptr = ptr.left
                else:
                    if ptr.right == None:
                        ptr.right = BgNode(d,None,None,None)
                    ptr = ptr.right
            if i < len(st)-1 and ptr.mid == None:
                ptr.mid = BgNode(st[i+1],None,None,None)
            if i < len(st)-1:
                ptr = ptr.mid
        ptr.mult += 1
        self.size += 1

    def addAll(self,A):
        for x in A: self.add(x)
        
    def printAll(self):
        def printFrom(ptr,s):
            if ptr == None: return
            s0 = s + ptr.data
            if ptr.left != None: printFrom(ptr.left,s)
            for i in range(ptr.mult): print(s0, end=" ")
            if ptr.mid != None: printFrom(ptr.mid,s+ptr.data)
            if ptr.right != None: printFrom(ptr.right,s)

        printFrom(self.root,""); print()      
        
    def count(self,st):
        # TODO: should count how many times st appears in the tree
        if st == "":
            return None
        ptr = self.root
        for i in range(len(st)):
            d = st[i]
            while True:
                #print(ptr)
                if ptr == None:
                    return 0
                if d == ptr.data:
                    if i < len(st) - 1:
                        ptr = ptr.mid
                    break
                elif d < ptr.data:
                    ptr = ptr.left
                elif d > ptr.data:
                    ptr = ptr.right
        return ptr.mult
    
    # the following recursive function adds in array A all elements in increasing order from node ptr on; 
    # s is the string read to reach node ptr, A the array to return, c the next position in A to fill
    def incRec(self,ptr,s):          #A,c): 
        # TODO
        if ptr == None:
            return []
        A = []                 # A will store the elements of the tree
        if ptr.left != None:
            A += self.incRec(ptr.left, s)
        d = ptr.data
        for _ in range(ptr.mult):
            A.append(s+d)
        if ptr.mid != None:
            A += self.incRec(ptr.mid, s+d)
        if ptr.right != None:
            A += self.incRec(ptr.right, s)
        return A    
    
    def toIncArray(self):
        # TODO: should return an array with all the elements in increasing order
        wordList = self.incRec(self.root, '')        
        return wordList

    # the following recursive function adds in array A all elements in decreasing order from node ptr on; 
    # s is the string read to reach node ptr, A the array to return, c the next position in A to fill        
    def decRec(self,ptr,s):          #A,c): 
        # TODO
        if ptr == None:
            return []
        A = []                 # A will store the elements of the tree
        if ptr.right != None:
            A += self.decRec(ptr.right, s)
        d = ptr.data
        if ptr.mid != None:
            A += self.decRec(ptr.mid, s+d)
        for _ in range(ptr.mult):
            A.append(s+d)
        if ptr.left != None:
            A += self.decRec(ptr.left, s)
        return A    
    
    def toDecArray(self):
        # TODO: should return an array with all the elements in decreasing order
        wordList = self.decRec(self.root, '')        
        return wordList
        
    def removeRec(self,ptr,s):
        if ptr == None:
            return None         
        d = s[0]
        if d == ptr.data:
            if len(s) == 1: 
                if ptr.mult > 0:
                    ptr.mult -= 1
            else:
                ptr.mid = self.removeRec(ptr.mid,s[1:])
        if d < ptr.data:   # and len(s) > 1:
            ptr.left = self.removeRec(ptr.left,s)
        if d > ptr.data:   # and len(s) > 1: 
            ptr.right = self.removeRec(ptr.right,s)
        if ptr.mult == 0 and ptr.mid == None:
            if ptr.left != None and ptr.right != None:
                inorderSucc = ptr.right
                while inorderSucc.left != None:
                    inorderSucc = inorderSucc.left
                inorderSucc.left = ptr.left
                tmp = ptr.right
                del ptr
                return tmp
            elif ptr.left != None and ptr.right == None:
                tmp = ptr.left
                del ptr
                return tmp
            elif ptr.left == None and ptr.right != None:
                tmp = ptr.right
                del ptr
                return tmp
            else:
                if ptr == self.root:
                    self.root = None
                del ptr
                return None
        return ptr
    
    def remove(self,st):
        # TODO: should remove one occurrence of st from the tree, removing also
        # any nodes that remain "dead" afterwards (i.e. with mult 0 and mid None) 
        if st == "":
            return None
        return self.removeRec(self.root,st)
```
