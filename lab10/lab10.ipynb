## Question 1

We define the Thribonacci sequence of numbers by the following function _thrib_:

- _thrib_(0) = 1
- _thrib_(1) = 1
- _thrib_(2) = 1
- _thrib_(n) = _thrib_(n-1) + 2 _thrib_(n-2) + 3 _thrib_(n-3) ,  if n > 2

Write a recursive Python function 

    def thrib(n)
    
that, on input `n`, returns _thrib_(`n`). 

Then, change your function into a dynamic programming one: 

    def thribDP(n)

using memoisation.

```python
def thrib(n):
  if n < 3:
    return 1
  return thrib(n-1) +2*(thrib(n-2)) + 3*(thrib(n-3))

def thribDPMemo(n, memo):

  if n < 3:
    return 1

  if n in memo:
    return memo[n]
  else:
    calc = thrib(n-1) +2*(thrib(n-2)) + 3*(thrib(n-3))
    memo[n] = calc
    return calc


def thribDP(n):
  memo = {}
  return thribDPMemo(n, memo)

print(thrib(20))
print(thribDP(20))
```

```python
def app(arr, v):
  A = [0 for _ in range(len(arr)+1)]
  for i in range(len(arr)):
    A[i] = arr[i]
  A[-1] = v
  return A

def maxBooksValDP2(w, bkWeight, bkVal):
    memo = {}
    inBagMemo = {}

    # Define the maxBooksMem function
    def maxBooksMem(w, bkWeight, bkVal, lo):
        # If the key (w, lo) exists in the memo dictionary, return the value at that key from the memo dictionary, and the indices of the books that are included in the bag in inBagMemo[w, lo]
        if (w,lo) in memo:
            return memo[w,lo], inBagMemo[w,lo]
        # If lo == len(bkWeight) or w < bkWeight[lo], set memo[w, lo] to 0, inBagMemo[w, lo] to an empty array, and return 0 and an empty array
        if lo == len(bkWeight) or w < bkWeight[lo]:
            memo[w,lo] = 0
            inBagMemo[w,lo] = []
            return 0, []
        # Otherwise, compute the maximum value of books that we can fill our bag with using only the first lo books in the bkWeight array, either by including the lo-th book or by not including it
        else:
            # If dp[i + 1] is equal to dp[i] + bkVal[i], this means that we include the lo-th book in our calculation, so we add its index to the indices of books that are included in the bag in inBagMemo[w, lo]
            withIt, withItInBag = maxBooksMem(w-bkWeight[lo],bkWeight,bkVal,lo+1)
            withoutIt, withoutItInBag = maxBooksMem(w,bkWeight,bkVal,lo+1)
            if withIt + bkVal[lo] > withoutIt:
                memo[w,lo] = withIt + bkVal[lo]
                # inBagMemo[w,lo] = withItInBag + [lo]
                inBagMemo[w, lo] = app(withItInBag, lo)
            # If dp[i + 1] is equal to dp[i], this means that we don't include the lo-th book in our calculation, so we keep the same indices of books that are included in the bag in inBagMemo[w, lo]
            else:
                memo[w,lo] = withoutIt
                inBagMemo[w,lo] = withoutItInBag
        # Return the value at the key (w, lo) in the memo dictionary, and the indices of the books that are included in the bag in inBagMemo[w, lo]
        return memo[w,lo], inBagMemo[w,lo]

    # Call the maxBooksMem function with initial values w, bkWeight, bkVal, and 0, and return the result
    return maxBooksMem(w, bkWeight, bkVal, 0)




  
bkWeight = [1,1,3,4,6,12,33,45,50]
bkVal = [1,2,5,1,10,20,24,5,60]
print(maxBooksValDP2(100,bkWeight,bkVal))
```

# Question 5

Write a Python function: 

    def closestSubset(s,A)

that takes an integer `s` and an array of positive integers `A` and returns an array consisting 
of elements of `A` which add up to `s`. If there is no subset that adds up to `s`, the function 
should instead return the subset which adds up to the value closest to `s`. 

For example: 

- if `A` is `[12, 79, 99, 91, 81, 47]` and `s` is `150`, it will return `[12, 91, 47]` as 12+91+47 is 150
- if `A` is `[15, 79, 99, 6, 69, 82, 32]` and `s` is `150` it will return `[69, 82]` as 69+82 is 151, and 
there is no subset of `A` whose sum is 150.

Test the method with arrays generated by `randomIntArray(s,n)` from Lab 3. Try with:

    A = randomIntArray(20,1000)
    subset = closestSubset(5000,A)

_Hint_: You can use a naive recursive solution for this problem. That is, to find the closest 
subset adding to `s` starting from position `i`, your algorithm should compare the best 
solution that includes `A[i]` with the best solution that does not include `A[i]`, and so on.
Such an algorithm is O(2<sup>n</sup>), with n the length of `A`, so it will take a significant amount of time
to process an array of length 20, and get overwhelmed for arrays much longer than that.

```python
import time, random

def randomIntArray(s,n):
  return [random.randint(0, n) for _ in range(s)]

def app(arr, v):
  A = [0 for _ in range(len(arr)+1)]
  for i in range(len(arr)):
    A[i] = arr[i]
  A[-1] = v
  return A

def closer(s, a, b):
  a_s = s - a
  b_s = s - b
  if a_s < 0: a_s = a_s*-1
  if b_s < 0: b_s = b_s*-1
  if a_s < b_s:
    return a
  else:
    return b

def closestSubset(s, A):

  def backtrack(i, sum, temp):
    if i == len(A) or sum == s:
      return sum, temp
    
    pick, x = backtrack(i+1, sum+A[i], app(temp, A[i]))
    leave, y = backtrack(i+1, sum, temp)

    if closer(s,pick,leave) == pick: return pick, x
    else: return leave, y

  return backtrack(0, 0, [])[1]

global ARR

ARR = randomIntArray(22, 1000)
t1 = time.time()
subset = closestSubset(5000,ARR)
# print(closestSubset(150,[12, 79, 99, 91, 81, 47]))
# print(closestSubset(150,[15, 79, 99, 6, 69, 82, 32]))
# print(closestSubset(8,[2, 3, 5, 7]))
print(closestSubset(8,[2,1,5]))
t2 = time.time()
print(t2-t1)
print(subset)
# print(closestSubset(8,[3,5]))

# 11.779114484786987
```

# Question 6

Using dynamic programming, write a Python function: 

    def closestSubsetDP(s,A)

that solves the same problem as Question 5 (but using dynamic programming this time). 
Test your function with random arrays as above. Does this version run faster than the 
simple recursive one?

```python
def closestSubsetDP(s, A):
  memo = {}

  def backtrack(i, sum, temp):
    if i == len(A) or sum == s:
      return sum
    
    if sum+A[i] not in memo: memo[sum+A[i]] = app(temp, A[i])
    a = backtrack(i+1, sum+A[i], memo[sum+A[i]])
    
    if sum not in memo: memo[sum] = temp
    b = backtrack(i+1, sum, temp)
    
    if closer(s,a,b) == a: 
      return a
    else: 
      return b

  return memo[backtrack(0, 0, [])]


t1 = time.time()
# print(closestSubsetDP(150,[12, 79, 99, 91, 81, 47]))
# print(closestSubsetDP(150,[15, 79, 99, 6, 69, 82, 32]))
# print(closestSubsetDP(8,[2, 3, 5, 7]))
# print(closestSubsetDP(8,[2,1,5]))
subset = closestSubsetDP(5000,ARR)
t2 = time.time()
print(t2-t1)
print(subset)

# 5.142787456512451
```
